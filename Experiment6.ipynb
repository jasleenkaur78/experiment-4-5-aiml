```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import files
upload = files.upload()
# Load the dataset
df = pd.read_csv('datasets.csv')
# Check column names and first few rows
print(df.columns)
print(df.head())

# Handle missing values
df['datasetName'].fillna(df['datasetName'].mode()[0], inplace=True)
df['about'].fillna(df['about'].mode()[0], inplace=True)
df['link'].fillna(df['link'].mode()[0], inplace=True)
df['categoryName'].fillna(df['categoryName'].mode()[0], inplace=True)
df['cloud'].fillna(df['cloud'].mode()[0], inplace=True)

# Fill missing values in 'vintage' column using the median
df['vintage'].fillna(df['vintage'].median(), inplace=True)

# Detect outliers in 'vintage' using IQR
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers

# Detect and print outliers in 'vintage'
outliers_vintage = detect_outliers_iqr(df, 'vintage')
print(f"Outliers in vintage: {len(outliers_vintage)}")

# Optionally, remove the outliers in 'vintage'
Q1 = df['vintage'].quantile(0.25)
Q3 = df['vintage'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers in 'vintage'
df = df[(df['vintage'] >= lower_bound) & (df['vintage'] <= upper_bound)]

# Verify the data after handling outliers
print(df['vintage'].describe())
```
